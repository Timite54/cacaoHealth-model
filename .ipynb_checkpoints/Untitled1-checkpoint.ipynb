```python
import pandas as pd
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import shutil

class CacaoDataProcessor:
    """
    Classe optimisée pour traiter les données du projet CacaoHealth, en se concentrant
    sur la détection du swollen shoot (CSSV) et la comparaison avec les plants sains
    """

    def __init__(self, base_dir='dataset', output_dir='dataset_processed'):
        self.base_dir = Path(base_dir)
        self.output_dir = Path(output_dir)

        # Configuration
        self.target_size = (224, 224)  # Taille standard pour de nombreux modèles CNN
        self.include_anthracnose = False  # Mettre à True pour inclure l'anthracnose

        # Chemins des fichiers et répertoires
        self.train_csv_path = self.base_dir / r'C:\Users\YEO\Python-Machine-Learning\Data\Train.csv'
        self.test_csv_path = self.base_dir / r'C:\Users\YEO\Python-Machine-Learning\Data\Test.csv'
        self.images_train_dir = self.base_dir / r'C:\Users\YEO\Python-Machine-Learning\Data\dataset\images\train'
        self.images_test_dir = self.base_dir / r'C:\Users\YEO\Python-Machine-Learning\Data\dataset\images\test'
        self.labels_train_dir = self.base_dir / r'C:\Users\YEO\Python-Machine-Learning\Data\dataset\labels\train'

        # Mappage des classes
        self.class_mapping = {
            'cssvd': 0,         # Swollen Shoot (CSSV)
            'healthy': 1,       # Plants sains
            'anthracnose': 2    # Autre maladie (si on décide de l'inclure)
        }

        # Création des répertoires de sortie
        self.create_output_directories()

    def create_output_directories(self):
        """Création des répertoires pour les données traitées"""
        # Répertoires principaux
        os.makedirs(self.output_dir, exist_ok=True)

        # Répertoires pour les données organisées par ensemble
        os.makedirs(self.output_dir / 'images' / 'train', exist_ok=True)
        os.makedirs(self.output_dir / 'images' / 'val', exist_ok=True)
        os.makedirs(self.output_dir / 'images' / 'test', exist_ok=True)

        # Répertoires pour les annotations
        os.makedirs(self.output_dir / 'annotations', exist_ok=True)

        # Répertoires séparés par classe pour la classification
        for disease in ['cssvd', 'healthy']:
            os.makedirs(self.output_dir / 'classification' / 'train' / disease, exist_ok=True)
            os.makedirs(self.output_dir / 'classification' / 'val' / disease, exist_ok=True)

        if self.include_anthracnose:
            os.makedirs(self.output_dir / 'classification' / 'train' / 'anthracnose', exist_ok=True)
            os.makedirs(self.output_dir / 'classification' / 'val' / 'anthracnose', exist_ok=True)

    def load_data(self):
        """Chargement des données CSV"""
        try:
            self.train_df = pd.read_csv(self.train_csv_path)
            self.test_df = pd.read_csv(self.test_csv_path)
            print(f"Données chargées: {len(self.train_df)} entrées d'entraînement, {len(self.test_df)} entrées de test")
            return True
        except Exception as e:
            print(f"Erreur lors du chargement des données: {e}")
            return False

    def explore_data(self):
        """Exploration des données pour comprendre leur structure"""
        print("\n=== EXPLORATION DES DONNÉES ===")

        print("\nStructure du fichier d'entraînement:")
        print(self.train_df.info())

        print("\nExemple de données d'entraînement:")
        print(self.train_df.head())

        # Analyse de la colonne de classe
        if 'class' in self.train_df.columns:
            self.class_column = 'class'
        else:
            potential_columns = [col for col in self.train_df.columns if 'class' in col.lower()]
            if potential_columns:
                self.class_column = potential_columns[0]
                print(f"Utilisation de '{self.class_column}' comme colonne de classe")
            else:
                print("Aucune colonne de classe trouvée")
                self.class_column = None
                return

        print(f"\nColonne identifiée pour les classes: '{self.class_column}'")
        print("\nDistribution des classes:")
        class_dist = self.train_df[self.class_column].value_counts()
        print(class_dist)

        # Identification de la colonne d'ID d'image
        if 'Image_ID' in self.train_df.columns:
            self.image_id_column = 'Image_ID'
        elif 'Image_id' in self.train_df.columns:
            self.image_id_column = 'Image_id'
        elif 'image_id' in self.train_df.columns:
            self.image_id_column = 'image_id'
        else:
            potential_columns = [col for col in self.train_df.columns if 'id' in col.lower() and 'image' in col.lower()]
            if potential_columns:
                self.image_id_column = potential_columns[0]
                print(f"Utilisation de '{self.image_id_column}' comme colonne d'identifiant d'image")
            else:
                print("Aucune colonne d'identifiant d'image trouvée")
                self.image_id_column = None

        # Vérifier si les coordonnées de boîtes englobantes sont disponibles (pour la détection d'objets)
        bbox_columns = ['xmin', 'ymin', 'xmax', 'ymax']
        self.has_bbox = all(col in self.train_df.columns for col in bbox_columns)

        if self.has_bbox:
            print("\nDétection de coordonnées de boîtes englobantes - préparation pour la détection d'objets")
        else:
            print("\nAucune coordonnée de boîte englobante détectée - préparation pour la classification")

    def prepare_data(self):
        """Prépare les données en filtrant les classes cibles"""
        if not self.class_column or not self.image_id_column:
            print("Colonnes de classe ou d'ID d'image non identifiées")
            return

        print("\n=== PRÉPARATION DES DONNÉES ===")

        # Filtrer les classes pertinentes
        target_classes = ['cssvd', 'healthy']
        if self.include_anthracnose:
            target_classes.append('anthracnose')

        mask = self.train_df[self.class_column].isin(target_classes)
        self.filtered_df = self.train_df[mask].copy()

        print(f"Données avant filtrage: {len(self.train_df)} entrées")
        print(f"Données après filtrage: {len(self.filtered_df)} entrées")

        # Distribution après filtrage
        print("\nDistribution après filtrage:")
        print(self.filtered_df[self.class_column].value_counts())

        # Vérifier l'équilibre des classes
        class_counts = self.filtered_df[self.class_column].value_counts()
        min_class_count = class_counts.min()
        max_class_count = class_counts.max()
        imbalance_ratio = max_class_count / min_class_count

        print(f"\nRatio de déséquilibre des classes: {imbalance_ratio:.2f}:1")
        if imbalance_ratio > 1.5:
            print("⚠️ Alerte: Déséquilibre significatif des classes détecté. Envisagez des techniques d'équilibrage.")

        # Ajouter une colonne pour l'ID numérique de classe
        self.filtered_df['class_num'] = self.filtered_df[self.class_column].map(self.class_mapping)

    def split_dataset(self, val_size=0.2, random_state=42):
        """Divise les données en ensembles d'entraînement et de validation en tenant compte des classes"""
        if not hasattr(self, 'filtered_df'):
            print("Exécutez d'abord prepare_data()")
            return

        print("\n=== DIVISION EN ENSEMBLES D'ENTRAÎNEMENT/VALIDATION ===")

        # Identifier les images uniques (pour éviter les doublons dus aux boîtes englobantes multiples)
        unique_images = self.filtered_df[self.image_id_column].unique()
        unique_df = self.filtered_df.drop_duplicates(subset=[self.image_id_column])

        # Division stratifiée par classe
        train_imgs, val_imgs = train_test_split(
            unique_images,
            test_size=val_size,
            random_state=random_state,
            stratify=unique_df[self.class_column] if len(unique_df) == len(unique_images) else None
        )

        # Création des dataframes d'entraînement et de validation
        self.train_df = self.filtered_df[self.filtered_df[self.image_id_column].isin(train_imgs)]
        self.val_df = self.filtered_df[self.filtered_df[self.image_id_column].isin(val_imgs)]

        print(f"Images uniques: {len(unique_images)}")
        print(f"Ensemble d'entraînement: {len(train_imgs)} images uniques, {len(self.train_df)} annotations")
        print(f"Ensemble de validation: {len(val_imgs)} images uniques, {len(self.val_df)} annotations")

        # Vérification de la distribution des classes
        print("\nDistribution des classes dans l'ensemble d'entraînement:")
        print(self.train_df[self.class_column].value_counts())

        print("\nDistribution des classes dans l'ensemble de validation:")
        print(self.val_df[self.class_column].value_counts())

    def process_and_organize_images(self):
        """Traite les images et les organise selon la structure requise"""
        if not hasattr(self, 'train_df') or not hasattr(self, 'val_df'):
            print("Exécutez d'abord split_dataset()")
            return

        print("\n=== TRAITEMENT ET ORGANISATION DES IMAGES ===")

        # Création d'une liste de toutes les images uniques
        train_images = set(self.train_df[self.image_id_column].unique())
        val_images = set(self.val_df[self.image_id_column].unique())

        # Traitement des images d'entraînement
        print("Traitement des images d'entraînement...")
        self._process_images(self.train_df, train_images, 'train')

        # Traitement des images de validation
        print("Traitement des images de validation...")
        self._process_images(self.val_df, val_images, 'val')

        # Préparation des annotations pour la détection d'objets
        if self.has_bbox:
            self._prepare_detection_annotations()

        # Préparation des images de test
        print("Préparation des images de test...")
        self._prepare_test_images()

    def _process_images(self, df, unique_images, split):
        """Traite les images pour un ensemble de données spécifique"""
        # Création d'un dictionnaire image_id -> classe pour une référence rapide
        image_class_map = df.drop_duplicates(subset=[self.image_id_column]).set_index(self.image_id_column)[self.class_column].to_dict()

        # Parcours des images uniques
        for img_id in tqdm(unique_images):
            # Détermination du chemin de l'image source
            img_filename = f"{img_id}"
            if not img_filename.endswith(('.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG')):
                img_filename += '.jpg'

            src_img_path = self.images_train_dir / img_filename

            # Vérifier différentes extensions si le fichier n'est pas trouvé
            if not os.path.exists(src_img_path):
                for ext in ['.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG']:
                    test_path = self.images_train_dir / (img_id.split('.')[0] + ext)
                    if os.path.exists(test_path):
                        src_img_path = test_path
                        break

            # Si l'image n'est pas trouvée, passer à la suivante
            if not os.path.exists(src_img_path):
                continue

            try:
                # Lecture de l'image
                img = cv2.imread(str(src_img_path))
                if img is None:
                    continue

                # Redimensionnement
                img_resized = cv2.resize(img, self.target_size)

                # Détermination de la classe
                img_class = image_class_map.get(img_id)
                if not img_class:
                    continue

                # Enregistrement pour l'ensemble d'entraînement/validation
                dst_path = self.output_dir / 'images' / split / f"{img_id}.jpg"
                cv2.imwrite(str(dst_path), img_resized)

                # Enregistrement pour la classification
                class_path = self.output_dir / 'classification' / split / img_class / f"{img_id}.jpg"
                cv2.imwrite(str(class_path), img_resized)

            except Exception as e:
                print(f"Erreur lors du traitement de l'image {img_id}: {e}")

    def _prepare_detection_annotations(self):
        """Prépare les annotations pour la détection d'objets"""
        print("\nPréparation des annotations pour la détection d'objets...")

        # Format YOLO: class_id center_x center_y width height (normalisés)
        for split, df in [('train', self.train_df), ('val', self.val_df)]:
            # Grouper par image_id pour traiter chaque image une seule fois
            for img_id, group in tqdm(df.groupby(self.image_id_column), desc=f"Annotations {split}"):
                # Chemin de l'image
                img_path = self.output_dir / 'images' / split / f"{img_id}.jpg"
                if not os.path.exists(img_path):
                    continue

                # Lire l'image pour obtenir ses dimensions
                img = cv2.imread(str(img_path))
                if img is None:
                    continue

                img_height, img_width = img.shape[:2]

                # Fichier d'annotation
                annotation_file = self.output_dir / 'annotations' / f"{img_id}.txt"

                with open(annotation_file, 'w') as f:
                    for _, row in group.iterrows():
                        if all(x in row for x in ['xmin', 'ymin', 'xmax', 'ymax']):
                            # Conversion au format YOLO
                            x_min, y_min = float(row['xmin']), float(row['ymin'])
                            x_max, y_max = float(row['xmax']), float(row['ymax'])

                            # Normalisation
                            center_x = ((x_min + x_max) / 2) / img_width
                            center_y = ((y_min + y_max) / 2) / img_height
                            width = (x_max - x_min) / img_width
                            height = (y_max - y_min) / img_height

                            # Classe (utilisez class_num si disponible, sinon mappez)
                            if 'class_num' in row:
                                class_id = int(row['class_num'])
                            else:
                                class_id = self.class_mapping.get(row[self.class_column], 0)

                            # Écriture de l'annotation
                            f.write(f"{class_id} {center_x:.6f} {center_y:.6f} {width:.6f} {height:.6f}\n")

    def _prepare_test_images(self):
        """Prépare les images de test"""
        if not hasattr(self, 'test_df'):
            print("Données de test non chargées")
            return

        # Identifier la colonne d'ID d'image dans les données de test
        test_img_col = self.image_id_column
        if test_img_col not in self.test_df.columns:
            # Recherche d'une colonne alternative
            for col in self.test_df.columns:
                if 'id' in col.lower() and ('image' in col.lower() or 'img' in col.lower                or 'file' in col.lower()):
                    test_img_col = col
                    print(f"Colonne d'ID d'image de test identifiée comme: '{test_img_col}'")
                    break
            if test_img_col not in self.test_df.columns:
                print("Impossible d'identifier la colonne d'ID d'image dans les données de test.")
                return

        print("Traitement des images de test...")
        test_images = self.test_df[test_img_col].unique()
        output_test_dir = self.output_dir / 'images' / 'test'
        os.makedirs(output_test_dir, exist_ok=True)

        for img_id in tqdm(test_images):
            img_filename = f"{img_id}"
            if not img_filename.endswith(('.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG')):
                img_filename += '.jpg'

            src_img_path = self.images_test_dir / img_filename

            if not os.path.exists(src_img_path):
                for ext in ['.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG']:
                    test_path = self.images_test_dir / (img_id.split('.')[0] + ext)
                    if os.path.exists(test_path):
                        src_img_path = test_path
                        break

            if os.path.exists(src_img_path):
                try:
                    img = cv2.imread(str(src_img_path))
                    if img is not None:
                        img_resized = cv2.resize(img, self.target_size)
                        dst_path = output_test_dir / f"{img_id}.jpg"
                        cv2.imwrite(str(dst_path), img_resized)
                except Exception as e:
                    print(f"Erreur lors du traitement de l'image de test {img_id}: {e}")

        print("Images de test préparées.")

    def visualize_data(self, num_samples=5):
        """Visualise quelques exemples d'images de chaque classe dans l'ensemble d'entraînement"""
        if not hasattr(self, 'train_df'):
            print("Exécutez d'abord split_dataset()")
            return

        print("\n=== VISUALISATION DES DONNÉES ===")
        if self.class_column is None:
            print("Colonne de classe non identifiée, la visualisation ne peut pas être effectuée par classe.")
            return

        unique_classes = self.train_df[self.class_column].unique()
        fig, axes = plt.subplots(len(unique_classes), num_samples, figsize=(15, 5 * len(unique_classes)))
        axes = axes.flatten()

        for i, cls in enumerate(unique_classes):
            class_df = self.train_df[self.train_df[self.class_column] == cls]
            sample_images = class_df[self.image_id_column].sample(min(num_samples, len(class_df)), random_state=42)

            for j, img_id in enumerate(sample_images):
                img_filename = f"{img_id}"
                if not img_filename.endswith(('.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG')):
                    img_filename += '.jpg'
                img_path = self.images_train_dir / img_filename

                # Vérifier différentes extensions
                if not os.path.exists(img_path):
                    for ext in ['.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG']:
                        test_path = self.images_train_dir / (img_id.split('.')[0] + ext)
                        if os.path.exists(test_path):
                            img_path = test_path
                            break

                if os.path.exists(img_path):
                    try:
                        img = cv2.imread(str(img_path))
                        if img is not None:
                            ax_index = i * num_samples + j
                            if ax_index < len(axes):
                                axes[ax_index].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
                                axes[ax_index].set_title(f"{cls}")
                                axes[ax_index].axis('off')
                    except Exception as e:
                        print(f"Erreur lors de la lecture de l'image {img_id}: {e}")

        plt.tight_layout()
        plt.show()

    def get_class_mapping(self):
        """Retourne le mappage des classes."""
        return self.class_mapping

    def get_output_directory(self):
        """Retourne le chemin du répertoire de sortie."""
        return str(self.output_dir)

def main():
    processor = CacaoDataProcessor()
    if processor.load_data():
        processor.explore_data()
        processor.prepare_data()
        processor.split_dataset()
        processor.process_and_organize_images()
        processor.visualize_data(num_samples=3) # Visualiser 3 exemples par classe
        print(f"\nDonnées traitées et organisées dans: {processor.get_output_directory()}")
        print(f"Mappage des classes: {processor.get_class_mapping()}")

if __name__ == "__main__":
    main()
              

      
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.applications import EfficientNetB0, ResNet50V2, MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Configuration pour améliorer les performances
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# Paramètres du modèle
BATCH_SIZE = 32
IMAGE_SIZE = (224, 224)  # Taille standard pour la plupart des modèles CNN
NUM_EPOCHS = 30
LEARNING_RATE = 1e-4
BASE_MODEL = 'efficientnet'  # Options: 'efficientnet', 'resnet', 'mobilenet'
FINE_TUNE_LAYERS = 30  # Nombre de couches à fine-tuner

class CacaoModel:
    def __init__(self, data_dir='dataset_processed', model_dir='models'):
        """
        Initialise le modèle pour la détection des maladies du cacaoyer
        
        :param data_dir: Répertoire contenant les données prétraitées
        :param model_dir: Répertoire de sauvegarde des modèles
        """
        self.data_dir = Path(data_dir)
        self.model_dir = Path(model_dir)
        os.makedirs(self.model_dir, exist_ok=True)
        
        # Chemins des données
        self.train_dir = self.data_dir / 'classification' / 'train'
        self.val_dir = self.data_dir / 'classification' / 'val'
        
        # Vérifier si les répertoires existent
        if not self.train_dir.exists() or not self.val_dir.exists():
            raise FileNotFoundError(f"Les répertoires {self.train_dir} ou {self.val_dir} n'existent pas")
        
        # Classes disponibles
        self.classes = [d.name for d in self.train_dir.iterdir() if d.is_dir()]
        self.num_classes = len(self.classes)
        print(f"Classes détectées: {self.classes}")
        
        # Générateurs de données
        self.train_generator = None
        self.val_generator = None
        
        # Modèle
        self.model = None
        self.history = None
    
    def setup_data_generators(self):
        """Configure les générateurs de données avec augmentation"""
        # Augmentation de données pour l'ensemble d'entraînement
        train_datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            vertical_flip=False,  # Les feuilles ont généralement une orientation
            fill_mode='nearest'
        )
        
        # Pas d'augmentation pour la validation, seulement la normalisation
        val_datagen = ImageDataGenerator(rescale=1./255)
        
        # Générateur d'entraînement
        self.train_generator = train_datagen.flow_from_directory(
            self.train_dir,
            target_size=IMAGE_SIZE,
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            shuffle=True
        )
        
        # Générateur de validation
        self.val_generator = val_datagen.flow_from_directory(
            self.val_dir,
            target_size=IMAGE_SIZE,
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            shuffle=False
        )
        
        print(f"Nombre d'échantillons d'entraînement: {self.train_generator.samples}")
        print(f"Nombre d'échantillons de validation: {self.val_generator.samples}")
        
        # Vérification des classes
        self.class_indices = self.train_generator.class_indices
        print(f"Indices des classes: {self.class_indices}")
    
    def build_model(self):
        """Construit le modèle de classification avec transfert d'apprentissage"""
        # Sélection du modèle de base
        if BASE_MODEL == 'efficientnet':
            base_model = EfficientNetB0(
                weights='imagenet',
                include_top=False,
                input_shape=(*IMAGE_SIZE, 3)
            )
        elif BASE_MODEL == 'resnet':
            base_model = ResNet50V2(
                weights='imagenet',
                include_top=False,
                input_shape=(*IMAGE_SIZE, 3)
            )
        elif BASE_MODEL == 'mobilenet':
            base_model = MobileNetV2(
                weights='imagenet',
                include_top=False,
                input_shape=(*IMAGE_SIZE, 3)
            )
        else:
            raise ValueError(f"Modèle {BASE_MODEL} non supporté")
        
        # Geler le modèle de base
        base_model.trainable = False
        
        # Construire le modèle complet
        model = Sequential([
            base_model,
            GlobalAveragePooling2D(),
            BatchNormalization(),
            Dense(256, activation='relu'),
            Dropout(0.5),
            BatchNormalization(),
            Dense(128, activation='relu'),
            Dropout(0.3),
            Dense(self.num_classes, activation='softmax')
        ])
        
        # Compiler le modèle
        model.compile(
            optimizer=Adam(learning_rate=LEARNING_RATE),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        model.summary()
        self.model = model
        return model
    
    def setup_callbacks(self):
        """Configure les callbacks pour l'entraînement"""
        # Checkpoint pour sauvegarder le meilleur modèle
        checkpoint = ModelCheckpoint(
            self.model_dir / 'cacao_best_model.h5',
            monitor='val_accuracy',
            save_best_only=True,
            mode='max',
            verbose=1
        )
        
        # Arrêt anticipé si pas d'amélioration
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=7,
            restore_best_weights=True,
            verbose=1
        )
        
        # Réduction du taux d'apprentissage si plateau
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=1e-6,
            verbose=1
        )
        
        return [checkpoint, early_stopping, reduce_lr]
    
    def train(self):
        """Entraîne le modèle en deux phases: gel puis fine-tuning"""
        if self.train_generator is None or self.val_generator is None:
            self.setup_data_generators()
        
        if self.model is None:
            self.build_model()
        
        callbacks = self.setup_callbacks()
        
        # Phase 1: Entraînement avec le modèle de base gelé
        print("Phase 1: Entraînement des couches supérieures...")
        history1 = self.model.fit(
            self.train_generator,
            validation_data=self.val_generator,
            epochs=10,
            callbacks=callbacks
        )
        
        # Phase 2: Fine-tuning avec dégel des dernières couches
        print("Phase 2: Fine-tuning des couches supérieures du modèle de base...")
        
        # On dégèle les dernières couches du modèle de base
        base_model = self.model.layers[0]
        base_model.trainable = True
        
        # Geler toutes les couches sauf les dernières
        for layer in base_model.layers[:-FINE_TUNE_LAYERS]:
            layer.trainable = False
        
        # Recompiler avec un taux d'apprentissage plus faible pour le fine-tuning
        self.model.compile(
            optimizer=Adam(learning_rate=LEARNING_RATE/10),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        # Résumé du modèle dégelé
        print("\nModèle après dégel pour fine-tuning:")
        self.model.summary()
        
        # Continuer l'entraînement avec fine-tuning
        history2 = self.model.fit(
            self.train_generator,
            validation_data=self.val_generator,
            epochs=NUM_EPOCHS - 10,
            callbacks=callbacks,
            initial_epoch=history1.epoch[-1] + 1
        )
        
        # Combiner les historiques
        self.history = {
            'accuracy': history1.history['accuracy'] + history2.history['accuracy'],
            'val_accuracy': history1.history['val_accuracy'] + history2.history['val_accuracy'],
            'loss': history1.history['loss'] + history2.history['loss'],
            'val_loss': history1.history['val_loss'] + history2.history['val_loss']
        }
        
        # Sauvegarder le modèle final
        self.model.save(self.model_dir / 'cacao_final_model.h5')
        print(f"Modèle sauvegardé dans {self.model_dir / 'cacao_final_model.h5'}")
        
        return self.history
    
    def evaluate(self):
        """Évalue le modèle sur l'ensemble de validation"""
        if self.model is None:
            raise ValueError("Le modèle n'a pas été entraîné")
        
        print("\nÉvaluation du modèle sur l'ensemble de validation...")
        
        # Prédiction sur l'ensemble de validation
        self.val_generator.reset()
        y_pred_proba = self.model.predict(self.val_generator)
        y_pred = np.argmax(y_pred_proba, axis=1)
        
        # Classe réelle
        y_true = self.val_generator.classes
        
        # Conversion des indices en noms de classe
        class_names = list(self.class_indices.keys())
        
        # Rapport de classification
        print("\nRapport de classification:")
        print(classification_report(y_true, y_pred, target_names=class_names))
        
        # Matrice de confusion
        cm = confusion_matrix(y_true, y_pred)
        
        # Affichage de la matrice de confusion
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", 
                    xticklabels=class_names, yticklabels=class_names)
        plt.title('Matrice de confusion')
        plt.ylabel('Vraie classe')
        plt.xlabel('Classe prédite')
        plt.tight_layout()
        plt.savefig(self.model_dir / 'confusion_matrix.png')
        
        # Affichage des courbes d'apprentissage
        if self.history:
            plt.figure(figsize=(12, 5))
            
            # Précision
            plt.subplot(1, 2, 1)
            plt.plot(self.history['accuracy'], label='Entraînement')
            plt.plot(self.history['val_accuracy'], label='Validation')
            plt.title('Précision')
            plt.xlabel('Époque')
            plt.ylabel('Précision')
            plt.legend()
            
            # Perte
            plt.subplot(1, 2, 2)
            plt.plot(self.history['loss'], label='Entraînement')
            plt.plot(self.history['val_loss'], label='Validation')
            plt.title('Perte')
            plt.xlabel('Époque')
            plt.ylabel('Perte')
            plt.legend()
            
            plt.tight_layout()
            plt.savefig(self.model_dir / 'learning_curves.png')
        
        # Calculer l'accuracy globale
        accuracy = accuracy_score(y_true, y_pred)
        print(f"\nPrécision globale: {accuracy:.4f}")
        
        return accuracy
    
    def visualize_model_predictions(self, num_samples=5):
        """Visualise quelques prédictions du modèle"""
        if self.model is None:
            raise ValueError("Le modèle n'a pas été entraîné")
        
        # Réinitialiser le générateur et obtenir un batch
        self.val_generator.reset()
        batch_x, batch_y = next(self.val_generator)
        
        # Prédictions
        pred_proba = self.model.predict(batch_x)
        pred_classes = np.argmax(pred_proba, axis=1)
        
        # Obtenir les noms des classes
        class_names = list(self.class_indices.keys())
        
        # Visualiser
        plt.figure(figsize=(15, 3*num_samples))
        for i in range(min(num_samples, len(batch_x))):
            plt.subplot(num_samples, 2, 2*i+1)
            plt.imshow(batch_x[i])
            plt.title(f"Vrai: {class_names[np.argmax(batch_y[i])]}")
            plt.axis('off')
            
            plt.subplot(num_samples, 2, 2*i+2)
            plt.bar(class_names, pred_proba[i])
            plt.title(f"Prédit: {class_names[pred_classes[i]]}")
            plt.xticks(rotation=45)
        
        plt.tight_layout()
        plt.savefig(self.model_dir / 'prediction_examples.png')
        
    def load_model(self, model_path):
        """Charge un modèle préalablement entraîné"""
        self.model = tf.keras.models.load_model(model_path)
        print(f"Modèle chargé depuis {model_path}")
        return self.model
    
    def predict_single_image(self, image_path):
        """Prédit la classe d'une seule image"""
        if self.model is None:
            raise ValueError("Le modèle n'a pas été chargé")
        
        # Charger et prétraiter l'image
        img = tf.keras.preprocessing.image.load_img(image_path, target_size=IMAGE_SIZE)
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) / 255.0
        
        # Prédiction
        prediction = self.model.predict(img_array)
        predicted_class_idx = np.argmax(prediction, axis=1)[0]
        
        # Inverser le dictionnaire des indices de classe
        idx_to_class = {v: k for k, v in self.class_indices.items()}
        predicted_class = idx_to_class[predicted_class_idx]
        
        # Probabilités
        class_probs = {idx_to_class[i]: float(prob) for i, prob in enumerate(prediction[0])}
        
        return {
            'class': predicted_class,
            'confidence': float(prediction[0][predicted_class_idx]),
            'probabilities': class_probs
        }


def main():
    # Créer et entraîner le modèle
    cacao_model = CacaoModel(data_dir='dataset_processed')
    cacao_model.setup_data_generators()
    cacao_model.build_model()
    history = cacao_model.train()
    
    # Évaluation du modèle
    accuracy = cacao_model.evaluate()
    cacao_model.visualize_model_predictions()
    
    print(f"\nEntraînement terminé avec une précision de {accuracy:.4f}")
    print(f"Les visualisations ont été sauvegardées dans {cacao_model.model_dir}")

if __name__ == "__main__":
    main()
```

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import cv2
import argparse
import time
from pathlib import Path

def load_and_preprocess_image(img_path, target_size=(224, 224)):
    """Charge et prétraite une image pour l'inférence"""
    # Vérifier si le chemin existe
    if not os.path.exists(img_path):
        raise FileNotFoundError(f"L'image {img_path} n'existe pas")
    
    # Charger l'image
    img = cv2.imread(img_path)
    if img is None:
        raise ValueError(f"Impossible de lire l'image {img_path}")
    
    # Conversion BGR à RGB (OpenCV charge en BGR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Redimensionnement
    img_resized = cv2.resize(img, target_size)
    
    # Normalisation
    img_normalized = img_resized / 255.0
    
    # Expansion des dimensions pour le batch (le modèle attend un batch)
    img_batch = np.expand_dims(img_normalized, axis=0)
    
    return img, img_batch

def predict_disease(model, img_batch, class_names):
    """Effectue une prédiction sur une image prétraitée"""
    # Mesure du temps d'inférence
    start_time = time.time()
    predictions = model.predict(img_batch)
    inference_time = time.time() - start_time
    
    # Obtenir la classe prédite et la confiance
    predicted_class_idx = np.argmax(predictions[0])
    confidence = float(predictions[0][predicted_class_idx])
    predicted_class = class_names[predicted_class_idx]
    
    # Créer un dictionnaire des probabilités par classe
    class_probabilities = {class_name: float(predictions[0][i]) 
                          for i, class_name in enumerate(class_names)}
    
    return {
        'class': predicted_class,
        'confidence': confidence,
        'probabilities': class_probabilities,
        'inference_time': inference_time
    }

def visualize_prediction(img, prediction, class_names, output_path=None):
    """Visualise l'image avec sa prédiction"""
    plt.figure(figsize=(12, 6))
    
    # Affichage de l'image
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title(f"Prédiction: {prediction['class']}\nConfiance: {prediction['confidence']:.2%}")
    plt.axis('off')
    
    # Affichage des probabilités
    plt.subplot(1, 2, 2)
    classes = list(prediction['probabilities'].keys())
    probs = list(prediction['probabilities'].values())
    
    # Tri décroissant des probabilités
    sorted_idx = np.argsort(probs)[::-1]
    classes = [classes[i] for i in sorted_idx]
    probs = [probs[i] for i in sorted_idx]
    
    colors = ['green' if c == prediction['class'] else 'blue' for c in classes]
    plt.barh(classes, probs, color=colors)
    plt.xlabel('Probabilité')
    plt.title(f"Temps d'inférence: {prediction['inference_time']:.3f} sec")
    plt.xlim(0, 1)
    plt.tight_layout()
    
    if output_path:
        plt.savefig(output_path)
        print(f"Visualisation sauvegardée dans {output_path}")
    else:
        plt.show()

def process_folder(model_path, folder_path, output_folder, class_names):
    """Traite toutes les images d'un dossier"""
    # Charger le modèle
    model = load_model(model_path)
    
    # Créer le dossier de sortie s'il n'existe pas
    os.makedirs(output_folder, exist_ok=True)
    
    # Liste des extensions d'image à traiter
    valid_extensions = ['.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG']
    
    # Obtenir la liste des fichiers image
    image_files = [f for f in os.listdir(folder_path) 
                   if os.path.isfile(os.path.join(folder_path, f)) and 
                   any(f.endswith(ext) for ext in valid_extensions)]
    
    results = []
    for img_file in image_files:
        img_path = os.path.join(folder_path, img_file)
        try:
            img, img_batch = load_and_preprocess_image(img_path)
            prediction = predict_disease(model, img_batch, class_names)
            
            # Créer un nom de fichier pour la sortie
            base_name = os.path.splitext(img_file)[0]
            output_path = os.path.join(output_folder, f"{base_name}_prediction.png")
            
            # Visualiser et sauvegarder
            visualize_prediction(img, prediction, class_names, output_path)
            
            results.append({
                'image': img_file,
                'prediction': prediction['class'],
                'confidence': prediction['confidence'],
                'output': output_path
            })
            
            print(f"Traité: {img_file} - Prédit: {prediction['class']} avec {prediction['confidence']:.2%} de confiance")
            
        except Exception as e:
            print(f"Erreur lors du traitement de {img_file}: {str(e)}")
    
    return results

def main():
    parser = argparse.ArgumentParser(description='Prédiction des maladies du cacaoyer')
    parser.add_argument('--model', type=str, required=True, help='Chemin vers le modèle entraîné (.h5)')
    parser.add_argument('--input', type=str, required=True, help='Chemin vers une image ou un dossier d\'images')
    parser.add_argument('--output', type=str, default='results', help='Dossier de sortie pour les visualisations')
    
    args = parser.parse_args()
    
    # Définir les classes (à adapter selon votre modèle)
    class_names = ['cssvd', 'healthy']  # Swollen Shoot et Sain
    
    # Vérifier si l'entrée est un fichier ou un dossier
    if os.path.isfile(args.input):
        # Traitement d'une seule image
        try:
            model = load_model(args.model)
            img, img_batch = load_and_preprocess_image(args.input)
            prediction = predict_disease(model, img_batch, class_names)
            
            # Créer le dossier de sortie
            os.makedirs(args.output, exist_ok=True)
            output_path = os.path.join(args.output, "prediction_result.png")
            
            visualize_prediction(img, prediction, class_names, output_path)
            
            print(f"Image analysée: {args.input}")
            print(f"Prédiction: {prediction['class']}")
            print(f"Confiance: {prediction['confidence']:.2%}")
            print(f"Temps d'inférence: {prediction['inference_time']:.3f} sec")
            
        except Exception as e:
            print(f"Erreur: {str(e)}")
            sys.exit(1)
    
    elif os.path.isdir(args.input):
        # Traitement d'un dossier d'images
        results = process_folder(args.model, args.input, args.output, class_names)
        print(f"\nTraitement terminé. {len(results)} images analysées.")
        print(f"Résultats sauvegardés dans: {args.output}")
    
    else:
        print(f"L'entrée {args.input} n'existe pas.")
        sys.exit(1)

if __name__ == "__main__":
    main()
```
